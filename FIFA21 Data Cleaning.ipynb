```python
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
# Define the function
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year
    return start_date, end_date, contract_length

# Applying the function to the DataFrame
new_data = fif['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

# Inserting new columns into the DataFrame
new_cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
for i in range(len(new_cols)):
    fif.insert(loc=fif.columns.get_loc('Contract')+1+i, column=new_cols[i], value=new_data[i])
```

```python
fif.sample(5)
```

```python
fif[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)']].sample(5)
```

```python
# contract categories

def contract_status(status):
    if status == 'Free':
        return 'Free'
    elif 'On Loan' in status:
        return 'On Loan'
    else:
        return 'Contract'

#add contract status column
fif.insert(fif.columns.get_loc('Contract Length(years)')+1, 'Contract Status', fif['Contract'].apply(contract_status))
```

## Height
### Player's weight in kilograms

```python
fif['Height'].dtype
```

```python
fif['Height'].unique()
```

```python
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)

# Apply Fn to height column
fif['Height'] = fif['Height'].apply(convert_height)
fif['Height'].unique()
```

```python
fif = fif.rename(columns={'Height':"Height(cm)"})
fif.sample(1)
```

## Weight
### Player's weight in kilograms

```python
fif['Weight'].dtype
```

```python
fif['Weight'].unique()
```

```python
def convert_weight(weight):
    if 'kg' in weight:
        return int(weight.strip('kg'))
    else:
        pounds = int(weight.strip('lbs'))
        return round(pounds/2.205)

# Apply fn to weight column
fif['Weight'] = fif['Weight'].apply(convert_weight)
fif['Weight'].unique()
```

```python
fif = fif.rename(columns={'Weight':"Weight(kg)"})
fif.sample(2)
```
